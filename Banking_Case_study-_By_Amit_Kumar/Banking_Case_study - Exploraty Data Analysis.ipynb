#### Recommended imputation strategy.

- AMT_GOODS_PRICE , AMT_ANNUITY - For these columns, it is suggested to impute them with the median. Since the data is numerical and contains outliers, using the median is advisable to avoid potential influence from outliers on the analysis, which might occur if the mean is used for imputation.

- EXT_SOURCE_2 - Imputing with the mean is recommended in this case since there is minimal difference between the median and mean, and the absence of significant outliers suggests that it won't impact the analysis significantly.

- NAME_TYPE_SUITE - Imputing these columns with the mode is suggested, given that they are categorical columns.

## Univariant Analysis On Categorical Ordered Data

```python
# Splitting Target variable
Target_1= app_data[app_data.TARGET==1]
Target_0= app_data[app_data.TARGET==0]
```

```python
def visualize_categorical(column):
    plt.style.use('ggplot')
    sns.despine

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

    sns.countplot(x=column, data=Target_0, ax=ax1)
    ax1.set_ylabel('Value Count')
    ax1.set_title(f'Distribution of {column} for Non-Defaulters (Target-0)\n', fontsize=13)
    ax1.set_xticklabels(ax1.get_xticklabels(), rotation=40, ha="right")

    for p in ax1.patches:
        ax1.annotate('{:.1f}%'.format((p.get_height() / len(Target_0)) * 100), (p.get_x() + 0.1, p.get_height() + 50))

    sns.countplot(x=column, data=Target_1, ax=ax2)
    ax2.set_ylabel('Value Count')
    ax2.set_title(f'Distribution of {column} for Defaulters (Target-1)\n', fontsize=12)
    ax2.set_xticklabels(ax2.get_xticklabels(), rotation=60, ha="right")

    for p in ax2.patches:
        ax2.annotate('{:.1f}%'.format((p.get_height() / len(Target_1)) * 100), (p.get_x() + 0.1, p.get_height() + 50))

    plt.show()
```

```python
## Insights on NAME_FAMILY_STATUS
visualize_categorical("NAME_FAMILY_STATUS")
```

Observations:

It is also noticeable that the percentage of defaulters among married individuals is elevated. Additionally, the observation reveals that there is a higher number of applicants who are married.

```python
## Insights on NAME_FAMILY_STATUS
visualize_categorical('CODE_GENDER')
```

Observations:

The graph above indicates a notable distribution among females in the non-defaulters group (Target_0), accounting for 66.6%. Similarly, in the defaulters group (Target_1), the distribution in females is relatively high at 57.1%, emphasizing that the percentage of non-defaulters is higher than that of defaulters.

```python
## Insights on NAME_INCOME_TYPE
visualize_categorical('NAME_INCOME_TYPE')
```

Observations:

Based on the preceding analysis, it is evident that the majority of defaulters, specifically 61.3%, are employed. Additionally, individuals who are employed constitute the largest group among non-payers.

```python
# Insights on NAME_EDUCATION_TYPE
visualize_categorical('NAME_EDUCATION_TYPE')
```

Observations:

The graph above illustrates that the highest number of applicants come from the group of individuals with secondary/secondary special education levels. Furthermore, there is a significant proportion of defaulters within this category.

```python
## Insights on NAME_HOUSING_TYPE
visualize_categorical('NAME_HOUSING_TYPE')
```

Observations::

The information depicted in the graph suggests that a greater number of applicants reside in apartments. Specifically, 85.7% of defaulters are individuals living in houses or apartments, whereas the percentage of non-defaulters residing in houses or apartments is 89%.

```python
## Specialized function for visualizing categorical variables
    
def plot_categorical_distribution(category_variable):

    plt.style.use('ggplot')
    sns.despine
    fig,ax = plt.subplots(1,1,figsize=(10,6))
    
    sns.countplot(x=category_variable,ax=ax,data=p_app_data,hue='NAME_CONTRACT_STATUS')
    plt.legend(loc='upper right')
    ax.set_ylabel(' Number of Application', fontsize= 15)
    ax.set_title(f'Distribution of {category_variable}\n',fontsize=15)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
    
    plt.show()
```

```python
## Correlation in Previous application data

num_col = p_app_data.select_dtypes(include=['int64', 'float64']).columns
corr_p_data = p_app_data[num_col].corr()

corr_p_data = corr_p_data.where(np.triu(np.ones(corr_p_data.shape), k=1).astype(bool))
corr_p_data = corr_p_data.unstack().reset_index()
corr_p_data.columns = ['Variable_1', 'Variable_2', 'Correlation']
corr_p_data.dropna(subset=['Correlation'], inplace=True)
corr_p_data['Correlation'] = round(corr_p_data['Correlation'], 2)

correlations_p_data = corr_p_data.sort_values(by='Correlation', ascending=False).head()
correlations_p_data
```

We can note a strong correlation among the following variables:

- AMT_GOODS_PRICE and AMT_APPLICATION
- AMT_GOODS_PRICE and AMT_CREDIT
- AMT_CREDIT and AMT_APPLICATION

```python
## Specialized Function for Bivariate Analysis on Categorical vs Numeric Columns

def cat_num_plot(category_column, numeric_column):

    plt.style.use('ggplot')
    sns.despine
    
    fig, ax = plt.subplots(1, 1, figsize=(8, 6))
    
    sns.boxplot(x=category_column, y=numeric_column, data=p_app_data)
    ax.set_ylabel(f'{numeric_column}')
    ax.set_xlabel(f'{category_column}')

    ax.set_title(f'{category_column} Vs {numeric_column}', fontsize=15)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")

    plt.show()
```

```python
## NAME_CONTRACT_STATUS status and AMT_ANNUITY
cat_num_plot('NAME_CONTRACT_STATUS', 'AMT_ANNUITY')
```

Observation:

The bivariate analysis reveals a notable increase in the "refused" category within the relationship of NAME_CONTRACT_STATUS and AMT_ANNUITY.

```python
## NAME_CONTRACT_STATUS and AMT_CREDIT
cat_num_plot('NAME_CONTRACT_STATUS', 'AMT_CREDIT')
```

Observation:

The bivariate analysis comparing NAME_CONTRACT_STATUS and AMT_CREDIT indicates a minimal percentage of unused offers and a significant increase in refusals.
